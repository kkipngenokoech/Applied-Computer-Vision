## Image Module

It provides the primary class for representing and manipulating images. It allows you to open, create, manipulate, and save images in various formats.

```python
from PIL import Image
```

## Common Attributes and Methods
### Attributes:
1. __image.format__: The file format of the image (e.g., JPEG, PNG).
2. __image.size__: The dimensions of the image as a tuple (width, height).
3. __image.mode__: The color mode of the image (e.g., RGB, RGBA, L).

### Methods:
1. __image.save(filename)__: Save the image to a file.
2. __image.show()__: Display the image using the default viewer.
3. __image.resize(size)__: Resize the image to the specified dimensions.
4. __image.rotate(angle)__: Rotate the image by the specified angle.
5. __image.crop(box)__: Crop the image to the specified region.
6. __image.convert(mode)__: Convert the image to a different mode (e.g., RGB to grayscale).
